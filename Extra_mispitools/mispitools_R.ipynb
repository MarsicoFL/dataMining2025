```r
sex_H1 <- LRsex("F", H = 1, LR = TRUE, seed = 123, nsims = 500)
sex_H2 <- LRsex("F", H = 2, LR = TRUE, seed = 123, nsims = 500)
col_H1 <- LRcol(1, H = 1, LR = TRUE, seed = 123, nsims = 500)
col_H2 <- LRcol(1, H = 2, LR = TRUE, seed = 123, nsims = 500)
age_H1 <- LRage(40, H = 1, LR = TRUE, seed = 123, nsims = 500)
age_H2 <- LRage(40, H = 2, LR = TRUE, seed = 123, nsims = 500)
```

```r
datasim <- simLR2dataframe(datasim)
combined_H1 <- datasimx$Related * sex_H1$LRs * col_H1$LRc * age_H1$LRa
combined_H2 <- datasimx$Unrelated * sex_H2$LRs * col_H2$LRc * age_H2$LRa
combined_datasim <- as.data.frame(cbind(combined_H2, combined_H1))
names(combined_datasim) <- c("Unrelated", "Related")
head(combined_datasim)
```

```r
DeT(combined_datasim, 10)
```
