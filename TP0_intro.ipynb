## Operaciones básicas

Las operaciones matemáticas en Python son intuitivas, lo que facilita cálculos relevantes en el contexto sanitario.

```python
# Operaciones aritméticas
peso = 25  # en kg
altura = 1.75  # en metros

# Cálculo del Índice de Masa Corporal (IMC)
imc = peso / (altura ** 2)

print(f"Peso: {peso} kg")
print(f"Altura: {altura} m")
print(f"IMC: {imc:.2f} kg/m²")

# Interpretación del IMC según criterios de la OMS
if imc < 18.5:
    categoria = "Bajo peso"
elif imc < 25:
    categoria = "Normopeso"
elif imc < 30:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidad"

print(f"Categoría: {categoria}")
```

## Funciones y modularización del código

Las funciones permiten encapsular operaciones recurrentes, mejorando la legibilidad y mantenimiento del código. Esto es particularmente valioso al implementar fórmulas médicas o protocolos de análisis estandarizados.

```python
def calcular_imc(peso, altura):
    """Calcula el Índice de Masa Corporal.

    Args:
        peso: Peso en kilogramos
        altura: Altura en metros

    Returns:
        float: Valor de IMC calculado
    """
    return peso / (altura ** 2)

def clasificar_imc(imc):
    """Clasifica el IMC según criterios de la OMS.

    Args:
        imc: Valor de IMC calculado

    Returns:
        str: Categoría correspondiente al IMC
    """
    if imc < 18.5:
        return "Bajo peso"
    elif imc < 25:
        return "Normopeso"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Ejemplo de aplicación con diferentes pacientes
pacientes = [
    {"nombre": "Laura Sánchez", "peso": 58.5, "altura": 1.62},
    {"nombre": "Roberto Gutiérrez", "peso": 85.0, "altura": 1.78},
    {"nombre": "Jimena Morales", "peso": 49.0, "altura": 1.60},
    {"nombre": "Eduardo Méndez", "peso": 92.3, "altura": 1.75}
]

print("Análisis antropométrico:")
print("{:<20} {:<10} {:<10} {:<10} {:<15}".format("Paciente", "Peso (kg)", "Altura (m)", "IMC", "Clasificación"))
print("-" * 65)

for paciente in pacientes:
    imc = calcular_imc(paciente["peso"], paciente["altura"])
    categoria = clasificar_imc(imc)
    print("{:<20} {:<10} {:<10} {:<10.2f} {:<15}".format(
        paciente["nombre"],
        paciente["peso"],
        paciente["altura"],
        imc,
        categoria
    ))
```

# 3. Análisis de datos con Pandas

La biblioteca Pandas constituye el pilar fundamental para el análisis de datos en Python. Su estructura principal, el DataFrame, facilita la manipulación de conjuntos de datos tabulares, como historias clínicas o registros epidemiológicos.

```python
# Importamos las bibliotecas necesarias
# %pip install pandas
import pandas as pd
import numpy as np

# Creamos un DataFrame con datos de pacientes
# (simulando un extracto de historia clínica electrónica)

# Generamos datos de ejemplo
np.random.seed(2)  # Para reproducibilidad
n_pacientes = 20

# Datos demográficos
nombres = ["Paciente_" + str(i) for i in range(1, n_pacientes + 1)]
edades = np.random.randint(18, 90, size=n_pacientes)
generos = np.random.choice(["M", "F"], size=n_pacientes)

# Datos clínicos
glucemia = np.random.normal(100, 25, n_pacientes).round(1)
presion_sistolica = np.random.normal(120, 15, n_pacientes).round().astype(int)
colesterol = np.random.normal(190, 30, n_pacientes).round().astype(int)
diagnosticos = np.random.choice(
    ["Hipertensión", "Diabetes tipo 2", "Dislipemia", "Cardiopatía", "Control de salud"],
    size=n_pacientes
)

# Creamos el DataFrame
datos = {
    "ID": nombres,
    "Edad": edades,
    "Género": generos,
    "Glucemia": glucemia,
    "Presión_sistólica": presion_sistolica,
    "Colesterol_total": colesterol,
    "Diagnóstico_principal": diagnosticos
}

df_pacientes = pd.DataFrame(datos)

# Mostramos los primeros registros
print("Registros clínicos (primeras 5 filas):")
df_pacientes.head()
```

```python
# Creamos columnas derivadas (ejemplo: clasificación de riesgo)
def clasificar_riesgo(fila):
    """Clasificación simplificada de riesgo cardiovascular."""
    # Esta es una clasificación simplificada para fines didácticos
    puntos = 0

    # Edad como factor de riesgo
    if fila["Edad"] > 65:
        puntos += 1

    # Glucemia como factor de riesgo
    if fila["Glucemia"] > 126:
        puntos += 1

    # Presión arterial como factor de riesgo
    if fila["Presión_sistólica"] > 140:
        puntos += 1

    # Colesterol como factor de riesgo
    if fila["Colesterol_total"] > 220:
        puntos += 1

    # Clasificación según puntos acumulados
    if puntos == 0:
        return "Bajo"
    elif puntos == 1:
        return "Moderado"
    elif puntos == 2:
        return "Alto"
    else:
        return "Muy alto"

# Aplicamos la función a cada fila
df_pacientes["Riesgo_CV"] = df_pacientes.apply(clasificar_riesgo, axis=1)

# Mostramos la distribución de riesgo
print("Distribución por nivel de riesgo cardiovascular:")
df_pacientes["Riesgo_CV"].value_counts()
```

# 5. Ejemplo de análisis sobre datos epidemiológicos

A continuación, realizaremos un análisis utilizando datos simulados de prevalencia de enfermedades respiratorias en diferentes regiones de Argentina, inspirado en patrones epidemiológicos reales.

```python
# Creamos datos simulados de prevalencia de enfermedades respiratorias por provincia
# Estos datos son ficticios pero inspirados en patrones epidemiológicos regionales

provincias = [
    "Buenos Aires", "CABA", "Catamarca", "Chaco", "Chubut", "Córdoba",
    "Corrientes", "Entre Ríos", "Formosa", "Jujuy", "La Pampa", "La Rioja",
    "Mendoza", "Misiones", "Neuquén", "Río Negro", "Salta", "San Juan",
    "San Luis", "Santa Cruz", "Santa Fe", "Santiago del Estero", "Tierra del Fuego", "Tucumán"
]

# Definimos regiones para análisis
region = [
    "Centro", "Centro", "NOA", "NEA", "Patagonia", "Centro",
    "NEA", "Centro", "NEA", "NOA", "Pampeana", "NOA",
    "Cuyo", "NEA", "Patagonia", "Patagonia", "NOA", "Cuyo",
    "Cuyo", "Patagonia", "Centro", "NOA", "Patagonia", "NOA"
]

# Generamos datos con variabilidad regional
np.random.seed(123)

# Tasas por 10,000 habitantes
asma = np.zeros(len(provincias))
epoc = np.zeros(len(provincias))
neumonia = np.zeros(len(provincias))
tuberculosis = np.zeros(len(provincias))

for i, r in enumerate(region):
    # Variación por región para simular factores ambientales y socioeconómicos
    if r == "NOA":
        # Mayor altitud, clima seco
        asma[i] = np.random.normal(60, 10)
        epoc[i] = np.random.normal(30, 8)
        neumonia[i] = np.random.normal(25, 5)
        tuberculosis[i] = np.random.normal(3.5, 1.0)
    elif r == "NEA":
        # Clima húmedo, tropical
        asma[i] = np.random.normal(70, 12)
        epoc[i] = np.random.normal(25, 6)
        neumonia[i] = np.random.normal(35, 8)
        tuberculosis[i] = np.random.normal(4.2, 1.2)
    elif r == "Centro":
        # Mayor urbanización
        asma[i] = np.random.normal(75, 15)
        epoc[i] = np.random.normal(35, 7)
        neumonia[i] = np.random.normal(28, 6)
        tuberculosis[i] = np.random.normal(2.5, 0.8)
    elif r == "Cuyo":
        # Clima árido
        asma[i] = np.random.normal(55, 9)
        epoc[i] = np.random.normal(38, 8)
        neumonia[i] = np.random.normal(22, 5)
        tuberculosis[i] = np.random.normal(2.8, 0.7)
    elif r == "Patagonia":
        # Clima frío, viento
        asma[i] = np.random.normal(50, 8)
        epoc[i] = np.random.normal(40, 10)
        neumonia[i] = np.random.normal(30, 7)
        tuberculosis[i] = np.random.normal(2.0, 0.6)
    else:  # Pampeana
        # Rural, exposición agrícola
        asma[i] = np.random.normal(65, 11)
        epoc[i] = np.random.normal(28, 6)
        neumonia[i] = np.random.normal(25, 5)
        tuberculosis[i] = np.random.normal(2.2, 0.6)

# Redondeamos a un decimal
asma = np.round(asma, 1)
epoc = np.round(epoc, 1)
neumonia = np.round(neumonia, 1)
tuberculosis = np.round(tuberculosis, 1)

# Creamos el DataFrame de datos epidemiológicos
datos_epidemio = {
    "Provincia": provincias,
    "Región": region,
    "Asma": asma,
    "EPOC": epoc,
    "Neumonía": neumonia,
    "Tuberculosis": tuberculosis
}

df_epidemio = pd.DataFrame(datos_epidemio)

# Mostramos los primeros registros
print("Datos epidemiológicos por provincia (casos por 10.000 habitantes):")
df_epidemio.head()

# Análisis por región
print("\nPromedios por región:")
df_epidemio.groupby("Región")[["Asma", "EPOC", "Neumonía", "Tuberculosis"]].mean()
```

## Aplicación de técnicas estadísticas para análisis epidemiológico

Realizaremos un análisis estadístico para evaluar si existen diferencias significativas en la prevalencia de enfermedades respiratorias entre regiones.

```python
from scipy import stats

# ANOVA para comparar asma entre regiones
regiones = df_epidemio["Región"].unique()
asma_por_region = [df_epidemio[df_epidemio["Región"] == region]["Asma"] for region in regiones]

# Realizamos el test ANOVA
f_stat, p_valor = stats.f_oneway(*asma_por_region)

print("Análisis de varianza (ANOVA) para tasas de asma entre regiones:")
print(f"Estadístico F: {f_stat:.4f}")
print(f"Valor p: {p_valor:.4f}")
print(f"Conclusión: {'Existen diferencias significativas' if p_valor < 0.05 else 'No hay diferencias significativas'} entre regiones (α=0.05)")

# Si el ANOVA es significativo, realizamos pruebas post-hoc
if p_valor < 0.05:
    from statsmodels.stats.multicomp import pairwise_tukeyhsd

    # Preparamos los datos para Tukey HSD
    asma_values = df_epidemio["Asma"]
    region_labels = df_epidemio["Región"]

    # Realizamos la prueba de Tukey
    tukey_result = pairwise_tukeyhsd(asma_values, region_labels, alpha=0.05)

    print("\nResultados de la prueba post-hoc de Tukey:")
    print(tukey_result)
```

# 6. Introducción al aprendizaje automático con scikit-learn

El aprendizaje automático ofrece herramientas poderosas para el análisis predictivo en salud. A continuación, exploraremos un ejemplo sencillo de clasificación para predecir el nivel de riesgo cardiovascular basado en los datos clínicos.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

# Preparamos los datos para el modelo
# Usaremos el DataFrame de pacientes creado anteriormente

# Variables predictoras (características)
X = df_pacientes[["Edad", "Glucemia", "Presión_sistólica", "Colesterol_total"]]

# Variable objetivo (etiqueta a predecir)
y = df_pacientes["Riesgo_CV"]

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Estandarizamos las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Creamos y entrenamos el modelo
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_scaled, y_train)

# Realizamos predicciones
y_pred = clf.predict(X_test_scaled)

# Evaluamos el rendimiento del modelo
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))

print("\nInforme de clasificación:")
print(classification_report(y_test, y_pred))
```

```python
# Importancia de las características
importancia = pd.DataFrame({
    'Característica': X.columns,
    'Importancia': clf.feature_importances_
}).sort_values('Importancia', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Característica', data=importancia)
plt.title('Importancia de las variables en la predicción del riesgo cardiovascular')
plt.tight_layout()
plt.show()
```

## Aplicación del modelo a nuevos datos

Demostraremos cómo utilizar el modelo entrenado para predecir el riesgo cardiovascular de nuevos pacientes.

```python
# Datos de nuevos pacientes
nuevos_pacientes = pd.DataFrame([
    {"Edad": 45, "Glucemia": 95, "Presión_sistólica": 125, "Colesterol_total": 190},  # Paciente 1
    {"Edad": 68, "Glucemia": 140, "Presión_sistólica": 150, "Colesterol_total": 230},  # Paciente 2
    {"Edad": 52, "Glucemia": 110, "Presión_sistólica": 142, "Colesterol_total": 210},  # Paciente 3
    {"Edad": 35, "Glucemia": 85, "Presión_sistólica": 118, "Colesterol_total": 175}    # Paciente 4
])

# Estandarizamos los datos usando el mismo escalador
nuevos_pacientes_scaled = scaler.transform(nuevos_pacientes)

# Realizamos predicciones
predicciones = clf.predict(nuevos_pacientes_scaled)

# Obtenemos las probabilidades para cada clase
probabilidades = clf.predict_proba(nuevos_pacientes_scaled)

# Añadimos los resultados al DataFrame
nuevos_pacientes["Riesgo_CV_Predicho"] = predicciones

# Mostramos los resultados
print("Predicciones para nuevos pacientes:")
print(nuevos_pacientes)

print("\nProbabilidades para cada categoría de riesgo:")
categorias_riesgo = clf.classes_
for i, paciente in enumerate(nuevos_pacientes.iterrows(), 1):
    print(f"Paciente {i}:")
    for j, categoria in enumerate(categorias_riesgo):
        print(f"  - {categoria}: {probabilidades[i-1][j]:.2%}")
```

# 7. Reflexiones sobre ética y privacidad en ciencia de datos de salud

El manejo de datos sanitarios conlleva importantes consideraciones éticas y legales que todo profesional del ámbito debe tener presentes.

## Aspectos clave en el manejo de datos de salud

1. **Confidencialidad y privacidad**
   - Los datos de salud están protegidos por la Ley de Protección de Datos Personales (Ley 25.326 en Argentina)
   - Se debe garantizar la anonimización efectiva de los datos
   - El consentimiento informado es fundamental para el uso de datos personales

2. **Equidad algoritmica**
   - Los modelos predictivos pueden perpetuar o amplificar sesgos existentes en los datos
   - Es fundamental evaluar el impacto diferencial de los algoritmos en distintos grupos poblacionales
   - La representatividad de los datos de entrenamiento es crucial para la generalización adecuada
